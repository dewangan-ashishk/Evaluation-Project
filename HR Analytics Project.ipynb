```python
x=df.drop(columns='Attrition',axis=1)
y=df['Attrition']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x.shape[1])]
vif['features']=x.columns
vif
```

The values of vif shows that ther is no multicolinearity in the features

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=7,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score, plot_roc_curve

from sklearn.model_selection import GridSearchCV
```

```python
models=[['Logistic Regression',LogisticRegression()],
       ['KNN Classifier',KNeighborsClassifier()],
       ['Decision Tree Classifier',DecisionTreeClassifier()],
       ['Random Forest Classifier',RandomForestClassifier()],
       ['XG Boost Classifier',XGBClassifier()]]
```

```python
eval=pd.DataFrame()
```

```python
for name,model in models:
    print('\n',name,'\n')
    model=model
    model.fit(x_train,y_train)
    prediction=model.predict(x_test)
    
    accuracy=accuracy_score(y_test,prediction)
    print(accuracy)
    
    conf_mat=confusion_matrix(y_test,prediction)
    print(conf_mat)
    
    report=classification_report(y_test,prediction)
    print(report)
```

Let us tune the hyper parameters and try to improve model performance before we finalize our model 

```python
algorithms={'KNN Classifier':{'model':KNeighborsClassifier(),
                             'param':{'algorithm':['kd_tree'],
                                      'leaf_size':[5,10,15,20,50,100,200],
                                      'n_neighbors':[3,5,7,9,13,17,19]}
                            },
           'Decison Tree Classifier':{'model':DecisionTreeClassifier(),
                                      'param':{'max_depth' : [3, 5, 7, 9, 11, 13, 15,20,25,50],
                                               'min_samples_leaf' : [3, 5, 7,13,15,20,50,100],
                                               'min_samples_split' : [3, 5, 7,13,15,20,50,100]}
                                     },
           'Random Forest Classifier':{'model':RandomForestClassifier(),
                                       'param':{'n_estimators':[5,10,15,20,25,50,100],
                                                'criterion':['entropy','gini'],
                                                'max_depth':[3,5,7,10,15,20,25,50,100],
                                                'min_samples_leaf':[3,5,7,10,15,20,25,50,100],
                                                'min_samples_split':[3,5,7,10,15,20,25,50,100]}
                                      },
           'XG Boost Classifier':{'model':XGBClassifier(),
                                  'param':{'max_depth': [3,5,7,10,13,15,20,25,50,100],
                                           'learning_rate': [0.01,0.03,0.05,0.1,0.3,0.5],
                                           'n_estimators': [5,10,15,20,25,50,100],
                                           'colsample_bytree': [0.3,0.5,0.7,0.9]}
                                 }
           }
```

```python
score=[]
for name,model_para in algorithms.items():
    rand_srch=RandomizedSearchCV(estimator=model_para['model'], 
                                 param_distributions=model_para['param'], 
                                 cv=10, n_jobs=-1, verbose=3)
    
    rand_srch.fit(x_train,y_train)
    score.append({'model':name,
                  'best_score':rand_srch.best_score_,
                  'best_parameters':rand_srch.best_params_
                 })
```

```python
model_performance=pd.DataFrame(score)
```

```python
model_performance
```

```python
print(model_performance['best_parameters'][0])
```

```python
#Checking the classfication report for KNN after hyper tuning
knn=KNeighborsClassifier(algorithm='kd_tree',n_neighbors=9,leaf_size=20)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
report=classification_report(y_test,y_pred_knn)
print(report)
```

```python
print(model_performance['best_parameters'][1])
```

```python
dt=DecisionTreeClassifier(min_samples_split=15, min_samples_leaf=13, max_depth=5)
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
report=classification_report(y_test,y_pred_dt)
print(report)
```

```python
print(model_performance['best_parameters'][2])
```

```python
rf=RandomForestClassifier(criterion='gini', n_estimators=5, min_samples_split=15, min_samples_leaf=5, max_depth=100)
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
report=classification_report(y_test,y_pred_rf)
print(report)
```

```python
print(model_performance['best_parameters'][3])
```

```python
xgb=XGBClassifier(n_estimators=50, colsample__bytree=0.7, max_depth=50, learning_rate=0.3)
xgb.fit(x_train,y_train)
y_pred_xgb=xgb.predict(x_test)
report=classification_report(y_test,y_pred_xgb)
print(report)
```

Out of all the models XG Boost Classifier of best with accuracy of 0.86 and precision of 0.83

```python
lr=LogisticRegression()
lr.fit(x_test,y_test)
disp=plot_roc_curve(lr,x_test,y_test)
plot_roc_curve(knn,x_test,y_test,ax=disp.ax_)
plot_roc_curve(dt,x_test,y_test,ax=disp.ax_)
plot_roc_curve(rf,x_test,y_test,ax=disp.ax_)
plot_roc_curve(xgb,x_test,y_test,ax=disp.ax_)
plt.legend(prop={'size':11},loc='lower right')
plt.show()
```

The AUC for Logistic Regression is maximum but the precision of XG Boost Classifier is maximum among all other models. Hence, we seelct XG Boost as our final model.
