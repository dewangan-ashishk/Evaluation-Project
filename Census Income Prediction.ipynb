```python
# Identify Categorical features
numerics=['int8','int16','int32','int64','float16','float32','float64']
categorical_columns=[]
features = df1.columns.values.tolist()

for col in features:
    if df1[col].dtype in numerics:
        continue
    categorical_columns.append(col)
```

Encoding Categorical Features

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
```

```python
df1['Workclass'] = encoder.fit_transform(df1['Workclass'])
df1['Education'] = encoder.fit_transform(df1['Education'])
df1['Marital_status'] = encoder.fit_transform(df1['Marital_status'])
df1['Occupation'] = encoder.fit_transform(df1['Occupation'])
df1['Relationship'] = encoder.fit_transform(df1['Relationship'])
df1['Race'] = encoder.fit_transform(df1['Race'])
df1['Sex'] = encoder.fit_transform(df1['Sex'])
df1['Native_country'] = encoder.fit_transform(df1['Native_country'])
```

```python
# Transforming the target variable to 0s if income is less than 50k and 1 if income is greater than 50k
df1["Income"]= df1["Income"].apply(lambda x: 0 if x == ' <=50K' else 1  )
```

Checking the Distribution Plot of Dataset

```python
plt.figure(figsize=(20,25))
plotnumber=1
for column in df1:
    if plotnumber<=15:#we won't print for region as it has categorical values
        ax = plt.subplot(3,5,plotnumber)
        sns.distplot(df1[column])
        plt.xlabel(column)
    plotnumber+=1
plt.tight_layout()
```

```python
# Divide dataset into features and labels
x=df1.drop(['Income'],axis=1)
y=df1['Income']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.25, random_state=7)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression().fit(x_train, y_train)
y_predict=log_reg.predict(x_test)
#pred_model2=model2.predict(X_test)
print("Logistic Regression Score : ",log_reg.score(x_test, y_test))
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report,accuracy_score, f1_score
cf=confusion_matrix(y_test, y_predict)
TP=cf[1][1]
TN=cf[0][0]
FN=cf[1][0]
FP=cf[0][1]
sns.heatmap(confusion_matrix(y_test, y_predict), annot=True, fmt='d')
plt.title("Logistic Regression Confusion Matrix")
```

```python
print('Accuracy:', accuracy_score(y_test,y_predict),"\n")
print('F1 score:', f1_score(y_test,y_predict),"\n")
precision=(TP/(TP+FP))
print("Precision:" , precision,"\n")
specificity=TN/(TN+FP)
print("Specificity:" , specificity,"\n")
recall=TP/(TP+FN)
print("Recall :" , recall,"\n")
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_predict)
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.plot(false_positive_rate, true_positive_rate, 'b',
label='AUC = %0.4f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.5])
plt.ylim([-0.1,1.5])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
print("ROC AUC Score : ", roc_auc_score(y_test, y_predict))
```
